**Battery Column** 

Divide into 2 columns battery capacity , Fastcharging 

5000 mAh Battery with 33W Fast Charging

```python
battery_capaciy=df1['battery'].str.strip().str.split(' ',n=2).str.get(0)
Fast_Charging=df1['battery'].str.strip().str.split('with').str.get(1).str.findall(r"\d{2,3}")
```

18,19,20 -> Position

```python
df1.insert(18,'Battery_Capaciy_(mAh)',value=battery_capaciy)
df1.insert(19,'Fast_Charging',value=Fast_Charging)
```

```python
df1['Battery_Capaciy_(mAh)']=df1['Battery_Capaciy_(mAh)'].str.replace('\u2009',' ').str.replace('mAh','').astype(float)
```

```python
def fast_charging_extractor(item):
    if type(item)==list:
        if len(item)==1:
            return item[0]
        else:
            return 0
    else:
        return -1
```

```python
df1['Fast_Charging']=df1['Fast_Charging'].apply(fast_charging_extractor)
```

Camera column will divide into 4 cols i.e. num_rear_camera , primary_rear_lens_in_MP , num_front_camera , primary_front_lens_in_MP

```python
def camera_extractor(MP):
    if 'Quad' in MP:
        return '4'
    elif 'Triple' in MP:
        return '3'
    elif 'Dual' in MP:
        return '2'
    elif 'Missing' in MP:
        return 'Missing'
    else:
        return '1'
```

```python
Num_Rear_Camera=df1['camera'].str.strip().str.split("&").str.get(0).apply(camera_extractor)
```

```python
Primary_Rear_Lens=df1['camera'].str.strip().str.split("Rear").str.get(0).str.replace('\u2009',' ').str.replace('Triple','').str.replace('Dual','').str.replace('Quad','').str.replace('+ Depth Sensor','').str.replace('Ultra Wide','').str.replace('Penta','').str.replace('Foldable Display,  Display','').str.strip()
```

```python
Num_Front_Camera=df1['camera'].str.strip().str.split("&").str.get(1).str.strip().fillna("Missing").apply(camera_extractor)
```

```python
Primary_Front_Lens=df1['camera'].str.strip().str.split("&").str.get(1).str.replace('\u2009',' ').str.replace('Front Camera','').str.replace('Dual','').str.strip().str.replace(' + TOF 3D','')
```

```python
df1.insert(25,'Num_Front_Camera',value=Num_Front_Camera)
df1.insert(26,'Primary_Front_Lens',Primary_Front_Lens)
df1.insert(27,'Num_Rear_Camera',value=Num_Rear_Camera)
df1.insert(28,'Primary_Rear_Lens',Primary_Rear_Lens)
```
